```python
import csv

# Specify the CSV file path
csv_file_path = "ganges.csv"



# Check if the CSV file already exists, if not, create it with headers
try:
    with open(csv_file_path, 'x', newline='') as csv_file:
        fieldnames = ["name","url","review_score","review_heading","profile_contr","review_body","date_of_stay","company"]
        writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
        writer.writeheader()
except FileExistsError:
    pass
```

```python
def extract_reviews(url, total):
    options.add_argument('--proxy-server=http://bf468f326d6ae92e70be440845041113f929d8ff:premium_proxy=true@proxy.zenrows.com:8001')
    # Create the webdriver with the specified options
    driver = webdriver.Chrome(options=options)
    count = 0
    driver.get(url)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    reviews = soup.find_all("div", class_="_c")
    del reviews[0]
    for review in reviews:
        count=count+1
        print("Review Number:",count)
        body = review.find("div", class_="biGQs _P pZUbB KxBGd")
        name = review.find("a", class_="BMQDV _F Gv wSSLS SwZTJ FGwzt ukgoS")
        url = review.find("div",class_="biGQs _P fiohW qWPrE ncFvv fOtGX")
        profile_contribution = review.find("div", class_="biGQs _P pZUbB osNWb")
        #print("profile: ",profile_contribution)
        contribution = profile_contribution.find_all("span")
        if len(contribution)>1:
            profile_contribution = contribution[1].text
        else:
            profile_contribution = contribution[0].text
        review_url = url.find("a", class_="BMQDV _F Gv wSSLS SwZTJ FGwzt ukgoS")['href']
        review_heading = url.find("span", class_='yCeTE')
        date = review.find("div", class_="biGQs _P pZUbB ncFvv osNWb")
        date_category = review.find("div", class_="RpeCd").text
        splitted = date_category.split("â€¢")
        if len(splitted)>1:
            travel_date=splitted[0]
            category=splitted[1]
        else:
            travel_date=splitted[0]
            category=" "
        #travel_date=date_category.split("â€¢")[0]
        #category = date_category.split("â€¢")[1]
        rating = review.find("svg", class_="UctUV d H0")['aria-label']
        final_url="https://www.tripadvisor.ca"+review_url
        print("URL: ",final_url)
        print("Reviewer name:",name.text)
        print("Profile Contribution:",profile_contribution)
        print("Review Heading:",review_heading.text)
        print("Review Body:",body.text)
        print("Review Date:",date.text)
        print("Rating:",rating)
        print("Travelled on and category:", travel_date)
        print("Category:",category)
        print("\n\n")
        with open('ganges.csv', 'a', encoding='utf-8', newline='') as csv_file:
            
            writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
            if csv_file.tell() == 0:
                # Write the header if the file is empty
                writer.writeheader()
            #writer.writeheader()
            writer.writerow({
                "name": name.text,
                "url": final_url,
                "review_score": rating,
                "review_heading": review_heading.text,
                "profile_contr": profile_contribution,
                "review_body": body.text,
                "date_of_stay": travel_date,
                "company": category
            })

            # Write the review data
            #writer.writerow({"name":name.text,"url":final_url,"review_score":rating,"review_heading":review_heading.text,"profile_contr":profile_contribution,"review_body":body.text,"date_of_stay":travel_date,"company":category})

        print("Review has been added to the CSV file.")
        
    total = total+count
    if(total==40):
        return
    next_page = soup.find_all("a", class_="BrOJk u j z _F wSSLS tIqAi unMkR")
    print(next_page, len(next_page))
    if len(next_page)>=2:
        next_url = 'https://www.tripadvisor.in'+next_page[1]['href']
        extract_reviews(next_url,total)
    else:
        next_url = 'https://www.tripadvisor.in'+next_page[0]['href']
        extract_reviews(next_url,total)
    
```
